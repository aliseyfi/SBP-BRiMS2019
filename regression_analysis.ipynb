### Regression Analysis on News Documents
In this notebook, we do all the regression analysis using the output of Coh-metrix on the collection of news documents.

```r
# reading the excel file which includes all the information
library("readxl")
my_data <- read_excel("data/FakeNewsNet/processed/fakenewsnet_full.xlsx")
head(my_data)
```

```r
# dropping the columns that we do not want in the regression analysis.
truth_labels <- my_data["label"]
# drops <- c("X__1", "DESPL", "label", "shares", "id")
drops <- c("X__1", "DESPL", "label", "id")
x <- my_data[, !(names(my_data) %in% drops)]
y <- my_data["shares"]
```

```r
# printing eigenvalues to find the number of components for PCA
# print(eigen(cor(x)))
```

```r
# doing PCA
# if you need to install the "psych" package, uncomment the two following lines
# options(download.file.method = "wget")
# install.packages("psych")
library("psych")
pca <- psych::principal(x, nfactors=22, rotate="varimax")
print(pca$loadings, cutoff = 0.4, sort = TRUE)
```

```r
# plotting the normal distribution and probability plot
# hist(unlist(y))
# qqnorm(unlist(y))
```

```r
# lin_model <- lm(unlist(y) ~ pca$scores + as.matrix(truth_labels))
# lin_model <- lm(unlist(y) ~ as.matrix(x) + as.matrix(truth_labels))
lin_model <- lm(shares ~ . + as.matrix(truth_labels), data=x)
# summary(lin_model)
tmp <- step(lin_model, direction = "backward")
summary(tmp)
```

```r
# fit a LASSO model
library(data.table)
library(glmnet)
library(ggplot2)
set.seed(123)
lasso_model = cv.glmnet(as.matrix(x), as.matrix(y), lambda=10^seq(4, -1, -.1), alpha=1)
lasso_model$lambda.1se
# plot(lasso_model)

best_lambda = lasso_model$lambda.1se
lasso_coef = lasso_model$glmnet.fit$beta[, lasso_model$glmnet.fit$lambda == best_lambda]
print(lasso_coef)
```
